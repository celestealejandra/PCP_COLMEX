### Ejercicio propuesto
Hacer un programa que pida al usuario su nombre, edad y su sexo, estatura, hobbie y los muestras de la siguiente forma </p>

</p> Te llamas: (nombre) </p>
</p> Tu edad es: (edad) </p>
</p> Eres: (sexo) </p>
</p> Mides: (estatura) </p>
</p> Hobbie: (Pasatiempo) </p>


```python
nombre = input("Introduce tu nombre: ")
edad = int(input("¿Cuántos años tienes? "))
sexo = input("¿Cómo te identificas? ")
estatura = float(input("¿Cuánto mides? "))
pasatiempo = input("Introduce un pasatiempo ")

name = "Te llamas: " + str(nombre) + " \n"
age = "Tienes: " + str(edad) + " años" +  " \n"
sex = "Eres: " + str(sexo) + " \n"
height = "Mides: " + str(estatura) + " \n"
hobbie = "Tu hobbie es: " + str(pasatiempo) + " \n"

print(name + age + sex + height + hobbie )


```

### Operadores aritméticos


```python
# Operadores ariméticos (+, -, *,. /, **, //)
```

```python
# Variables
n1 = 10
n2 = 7
```

```python
# Suma
res = n1 + n2
print(res)
```

```python
# Resta
res = n1 - n2
print(res)
```

```python
# Multiplicación
res = n1 * n2
print(res)
```

```python
# División
res = n1 / n2
print(res)

# Otras variables
n3 = 20
n4 = 15
res = n3 / n4
print(res)
```

```python
# Modulo - residuo
res = n1 % n2
print(res)

res = n3 % n4
print(res)
```

```python
# División entera
res = n1 // n2
print(res)

res = n3 // n4
print(res)
```

```python
# Potencia
res = n1 ** n2
print(res)

res = n3 ** n4
print(res)
#
```
